```python
import pandas as pd
import data_cleaner as clean
```

```python
#Load datasets
who_1 = pd.read_csv("../data/WHO-CHILD-HEALTH-1.csv")
who_2 = pd.read_csv("../data/WHO-CHILD-HEALTH-2a.csv")
who_3 = pd.read_csv("../data/WHO-CHILD-HEALTH-3.csv")
who_4 = pd.read_csv("../data/WHO-CHILD-HEALTH-4.csv")
who_maternal = pd.read_csv("../data/WHO-MATERNAL-HEALTH_Variables.csv")
who_sanitation_1 = pd.read_csv("../data/WHO-SANITATION-1.csv")
who_sanitation_2 = pd.read_csv("../data/WHO-SANITATION-2.csv")
who_substance = pd.read_csv("../data/WHO-SUBSTANCE-USE-1.csv")
who_variables = pd.read_csv("../data/WHO_IMR_Variables.csv")
imr_genders = pd.read_csv("../data/WHO-IMR-ALL-YEARS-BY-GENDER.csv")
wb_gdp = pd.read_csv("../data/WorldBank_GDP_Data.csv", header=2)
who_extras = pd.read_csv("../data/WHO-EXTRAS.csv")
#wb_gdp = wb_gdp.set_index('Country Code')
```

```python
#Filter imr so only both genders included
imr_both = imr_genders[imr_genders['SEX'] == "BTSX"]
imr_both['SEX'].value_counts()
#imr_both.head()
```

```python
#Filter and Organize WHO 1
who_1['Country_Year'] = who_1['COUNTRY'].str.cat(who_1['YEAR'].astype(str),sep="_")
who_1 = who_1.set_index('Country_Year')

who_1_pivot = pd.pivot_table(who_1, index=['Country_Year'], columns='GHO', values='Display Value')
```

```python
#Filter and Organize WHO 3
who_3['Country_Year'] = who_3['COUNTRY'].str.cat(who_3['YEAR'].astype(str),sep="_")
who_3 = who_3.set_index('Country_Year')

#Only leave both sexes or unspecified
who_3_filtered = who_3[~who_3['SEX'].isin(['MLE','FMLE'])]
#Keep only useful datasets (that had data or matched requirements for whole country data)
who_3_filtered = who_3_filtered[who_3_filtered['GHO'].isin(['NUTRITION_564','WHS4_128','MCV2'])]

who_3_pivot = pd.pivot_table(who_3_filtered, index=['Country_Year'], columns='GHO', values='Numeric')
```

```python
#Filter and Organize WHO 4
who_4['Country_Year'] = who_4['COUNTRY'].str.cat(who_4['YEAR'].astype(str),sep="_")
who_4 = who_4.set_index('Country_Year')

#Keep only useful datasets (that had data or matched requirements for whole country data)
who_4_filtered = who_4[~who_4['GHO'].isin(['NUTUNDERWEIGHTPREV'])]

who_4_pivot = pd.pivot_table(who_4_filtered, index=['Country_Year'], columns='GHO', values='Numeric')
```

```python
#Filter and Organize WHO maternal factor data
who_maternal['Country_Year'] = who_maternal['COUNTRY'].str.cat(who_maternal['YEAR'].astype(str),sep="_")
who_maternal = who_maternal.set_index('Country_Year')

who_maternal_pivot = pd.pivot_table(who_maternal, index=['Country_Year'], columns='GHO', values='Numeric')
```

```python
#Filter and Organize WHO Sanitation data 1
who_sanitation_1['Country_Year'] = who_sanitation_1['COUNTRY'].str.cat(who_sanitation_1['YEAR'].astype(str),sep="_")
who_sanitation_1 = who_sanitation_1.set_index('Country_Year')

who_sanitation_1_pivot = pd.pivot_table(who_sanitation_1, index=['Country_Year'], columns='GHO', values='Numeric')
```

```python
#Filter and Organize WHO Sanitation data 2
who_sanitation_2['Country_Year'] = who_sanitation_2['COUNTRY'].str.cat(who_sanitation_2['YEAR'].astype(str),sep="_")
who_sanitation_2 = who_sanitation_2.set_index('Country_Year')

#Keep only useful datasets (that had data or matched requirements for whole country data)
who_sanitation_2_filtered =  who_sanitation_2[who_sanitation_2['RESIDENCEAREATYPE'].isin(['TOTL'])]

who_sanitation_2_pivot = pd.pivot_table(who_sanitation_2_filtered, index=['Country_Year'], columns='GHO', values='Numeric')
```

```python
#Filter and Organize WHO Substance Abuse data
who_substance['Country_Year'] = who_substance['COUNTRY'].str.cat(who_substance['YEAR'].astype(str),sep="_")
who_substance = who_substance.set_index('Country_Year')

#This is too look at maternal tabaco consumption, so we are filtering for females
who_substance_filtered =  who_substance[who_substance['SEX'].isin(['FMLE'])]

who_substance_pivot = pd.pivot_table(who_substance_filtered, index=['Country_Year'], columns='GHO', values='Numeric')
```

```python
#Filter and Organize other WHO data
who_variables['Country_Year'] = who_variables['COUNTRY'].str.cat(who_variables['YEAR'].astype(str),sep="_")
who_variables = who_variables.set_index('Country_Year')

who_variables_pivot = pd.pivot_table(who_variables, index=['Country_Year'], columns='GHO', values='Numeric')
```

```python
#Filter and Organize extra WHO data
who_extras['Country_Year'] = who_extras['COUNTRY'].str.cat(who_extras['YEAR'].astype(str),sep="_")
who_extras = who_extras.set_index('Country_Year')

who_extras_pivot = pd.pivot_table(who_extras, index=['Country_Year'], columns='GHO', values='Numeric')
```

```python
#Filter and Organize World Bank GDP data
wb_gdp
wb_gdp_drop = wb_gdp.drop(['Country Name', 'Indicator Name', 'Indicator Code', 'Unnamed: 65'], axis=1)
wb_gdp_melted = pd.melt(wb_gdp_drop, id_vars="Country Code", var_name="Year",value_name="GDP")

wb_gdp_melted['Country_Year'] = wb_gdp_melted['Country Code'].str.cat(wb_gdp_melted['Year'].astype(str),sep="_")
wb_gdp_melted = wb_gdp_melted.set_index('Country_Year')
```

```python
#Filter and Organize IMR data
imr_both['Country_Year'] = imr_both['COUNTRY'].str.cat(imr_both['YEAR'].astype(str),sep="_")
imr_both = imr_both.set_index('Country_Year')

imr_pivot = pd.pivot_table(imr_both, index=['Country_Year'], columns='GHO', values='Numeric')
```

```python
#Join the various data frames into one table
to_join = [who_1_pivot,who_3_pivot,who_4_pivot,who_maternal_pivot,who_sanitation_1_pivot,who_sanitation_2_pivot,who_substance_pivot,who_variables_pivot,wb_gdp_melted,who_extras_pivot]

simple_joined_data = imr_pivot

for pivoted in to_join:
    simple_joined_data = simple_joined_data.join(pivoted, on=["Country_Year"], how="left", lsuffix="DROP").filter(regex="^(?!.*DROP)")

simple_joined_data
#simple_joined_data.count(0)
```

```python
#Write csv
simple_joined_data.to_csv("../data/WHO-Weston-Fun.csv")
```
