## Binarize columns and replace NaN with "Not Specified"

```python
replacement_string = 'Not_Specified'
```

```python
cols_to_encode = [
    'EDUCATIONLEVEL', 
    'SEX', 
    'RESIDENCEAREATYPE', 
    'WEALTHQUINTILE',
    'WEALTHDECILE'
]

for col in cols_to_encode:
    imr_basic[col] = imr_all[col].fillna(replacement_string) 
```
